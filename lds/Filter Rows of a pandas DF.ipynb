## building up with a multi-step process

```python
type(True)
```

```python
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[:5]
```

```python
len(booleans)
```

```python
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long]
```
